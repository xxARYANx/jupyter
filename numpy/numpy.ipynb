### array creation

```python
import numpy as np
```

```python
arr1=np.array([2,"j"])
print(arr1)
```

```python
arr2=np.array([[1,2],[2,3]])
print(arr2)
```

```python
arr3=np.zeros((3,4))
arr3
```

```python
arr4=np.ones((2,3))
print(arr4)
```

```python
arr5=np.identity(5)
arr5
```

```python
print(np.arange(5,10))
```

```python
arr9=np.linspace(10,20,15)
print(arr9)
```

```python
arr0=arr9.copy()
print(arr0)
print(np.ndim(arr0))
arr0.shape
arr0.ndim
```

```python
arr3=np.array([[[1,2],[3,4]],[[4,5],[6,7]]])
print(arr3)
print(arr3.shape)
print(arr3.ndim)
print(arr3.size)
print(arr3.dtype)
print(arr3.astype('float'))
```

```python
ar=np.arange(24).reshape(6,4)
print(ar)
print()
print(ar[:,2])
print(ar[2:4,1:3])
print(ar[4:,2])
```

```python
ar=np.arange(24).reshape(6,4)
for i in ar:
    print(i)
for i in np.nditer(ar):
    print(i)
```

## operations on array

```python
arr1=np.arange(20,30).reshape(2,5)
print(arr1)
arr2=np.arange(10,20).reshape(2,5)
print(arr2)
print (arr1-arr2)

print(arr1>25)

print(arr1* arr2)
print(arr1*2)
print(arr1.max())
print(arr1.min())
```

```python
a=np.array([[1,2,3],[4,5,6]])
# b=np.array([[7,8,9],[10,12,14]])
b=np.arange(7,13).reshape(3,2)
print(a.dot(b))
print(a)
print()
print(a.max(axis=0))
print()
print(a.max(axis=1))
```

```python
arr4=np.arange(10,20)
print(arr4.mean())
print(np.mean(arr4))
print(arr4.std())
```
