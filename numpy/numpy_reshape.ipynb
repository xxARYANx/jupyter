```python
import numpy as np
```

```python
a=np.arange(10,22).reshape(3,4)
print(a)
print(a.ravel()) #this function create any dimension array into 1d array
print(a)
print()
print(a.transpose())
```

```python
arr1=np.arange(1,9).reshape(4,2)
print(arr1)
arr2=np.arange(1,9).reshape(4,2)
print(arr1)
print()
print(np.hstack((arr1,arr2)))
print()
print(np.vstack((arr1,arr2)))
```

```python
arr3=np.random.randint(1,100,20).reshape(5,4)
print(arr3)
# to filter something from array we call it as BOOLEAN INDEXING
print(arr3[arr3>30])
```
