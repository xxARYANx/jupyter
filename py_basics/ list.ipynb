```python
l=[1,2,3,2,1,3,1]
l[:3]=[0,11,22]
print(l)
l.append(1000)
print(l)
l.extend(['sexa','me'])
print(l)
l.append("goa")
print(l)
l.extend("bihar")
print(l)
l.insert(0,"list hu bey")
print(l)
print(len(l))
```

### delete operations in list

```python
l=['list hu bey', 0, 11, 22, 2, 1, 3, 1, 1000, 'sexa', 'me', 'goa', 'b', 'i', 'h', 'a', 'r']
print(l)
l.remove(0)
print(l)
print(l.pop())
del l[0]
print(l)
l.clear()
print(l)
```
