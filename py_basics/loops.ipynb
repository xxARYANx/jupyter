## for loops

```python
for i in range(1,10):
    print(i)
```

```python
for i in range(1,10):
    for j in range(0,i):
        print(j,end='')
    print(" ")
```

```python
i=1
num=int(input("enter the number you want the table for:"))
while i<=10:
    print(i*num)
    i=i+1
```

```python
num=10
for i in range (1,11):
    print(num*i)
```

```python
fruits=["mango","apple","banana"]
for x in fruits:
    print(x)
```
