```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
 batsman=pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('kohli vs sharma career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='yellow')
plt.plot(batsman['index'],batsman['RG Sharma'],color='m')
plt.title('kohli vs sharma career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dotted',linewidth=1)
plt.plot(batsman['index'],batsman['RG Sharma'],color='m',linestyle='dashed',linewidth=2)
plt.title('kohli vs sharma career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dotted',linewidth=1,marker='o',markersize=5)
plt.plot(batsman['index'],batsman['RG Sharma'],color='m',linestyle='dashed',linewidth=2)
plt.title('kohli vs sharma career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='green',linestyle='dotted',linewidth=1,marker='o',markersize=5,label='virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='m',linestyle='dashed',linewidth=2,label='rohit')
plt.title('kohli vs sharma career comparison')
plt.xlabel('season')
plt.ylabel('runs scored')
#for displaying label
plt.legend()
#for defining axis limit
plt.xlim(2008,2015)
plt.ylim(200,600)
#for displaying grids
plt.grid()
#for displaying in code editor we generally use show function
plt.show()
```
