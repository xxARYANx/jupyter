```python
#we can create a function also where toss decision is field like:
def team_match(team1,team2):
    mask=(data['toss_decision']=="field")&((data['team1']==team1)|(data['team2']==team2))
    return data[mask]
team_match('Gujrat Lions','Royal Challengers Bangalore')
```

```python
all_batsman=mask.groupby('batsman')['batsman_runs'].count()
x=all_batsman>200
all_batsman[x]
```

```python
batsman_list=all_batsman[x].index.tolist()
```

```python
final=mask[mask['batsman'].isin(batsman_list)]
```

```python
final.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False)
```
